```python
import json
import os
from gensim.corpora import Dictionary
from gensim.models import LdaModel
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk

# Download NLTK data
nltk.download('punkt')
nltk.download('stopwords')

# Function to load JSONL file
def load_jsonl(file_path):
    data = []
    with open(file_path, 'r') as f:
        for line in f:
            data.append(json.loads(line.strip()))
    return data

# Function to filter data based on evidence_query
def filter_data(data, query_filter):
    filtered_sentences = []
    for item in data:
        evidences = item.get('meta', {}).get('evidences', [])
        for evidence_group in evidences:
            for evidence in evidence_group:
                if evidence.get('evidence_query', '').lower() == query_filter.lower():
                    filtered_sentences.append(evidence.get('evidence_extract_from_source', ''))
    return filtered_sentences

# Function to preprocess text
def preprocess_text(text):
    stop_words = set(stopwords.words('english'))
    tokens = word_tokenize(text)
    tokens = [word.lower() for word in tokens if word.isalpha() and word not in stop_words]
    return tokens

# Function to perform topic modeling
def topic_modeling(texts, num_topics=5, num_words=10):
    # Preprocess texts
    processed_texts = [preprocess_text(text) for text in texts]

    # Print preprocessed texts for debugging
    print("Preprocessed Texts:")
    for text in processed_texts[:5]:  # Print the first 5 for inspection
        print(text)

    # Create a dictionary and corpus
    dictionary = Dictionary(processed_texts)
    corpus = [dictionary.doc2bow(text) for text in processed_texts]

    # Print dictionary and corpus for debugging
    print("\nDictionary:")
    print(dictionary.token2id)
    print("\nCorpus:")
    for doc in corpus[:5]:  # Print the first 5 for inspection
        print(doc)

    # Train LDA model
    lda = LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=15)

    # Print topics
    topics = lda.print_topics(num_words=num_words)
    for topic in topics:
        print(topic)

# Mount Google Drive
drive.mount('/content/drive')

# Replace 'path_to_your_directory' with the actual path
path = '/content/drive/My Drive/lobbymap_dataset'

# Replace 'your_sentence_level_jsonl_file.jsonl' with the actual file name
jsonl_file_path = os.path.join(path, 'train.jsonl')

# Define the query you want to filter by
#query_filter = "Alignment with IPCC on Climate Action"  # Replace with your desired query
# query_filter = "Carbon Tax"
query_filter = "GHG Emission Regulation"


# Load and filter the JSONL file based on the query
data = load_jsonl(jsonl_file_path)
filtered_sentences = filter_data(data, query_filter)

# # Print filtered sentences for debugging
# print("\nFiltered Sentences:")
# for sentence in filtered_sentences[:5]:  # Print the first 5 for inspection
#     print(sentence)

# Perform topic modeling
print("\nTopic Modeling Results:")
topic_modeling(filtered_sentences)
```

```python
import json
import os
import shutil
from gensim.corpora import Dictionary
from gensim.models import LdaModel
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
import pyLDAvis.gensim_models as gensimvis
import pyLDAvis

# Download NLTK data
nltk.download('punkt')
nltk.download('stopwords')

# Function to load JSONL file
def load_jsonl(file_path):
    data = []
    with open(file_path, 'r') as f:
        for line in f:
            data.append(json.loads(line.strip()))
    return data

# Function to filter data based on evidence_query
def filter_data(data, query_filter):
    filtered_sentences = []
    for item in data:
        evidences = item.get('meta', {}).get('evidences', [])
        for evidence_group in evidences:
            for evidence in evidence_group:
                if evidence.get('evidence_query', '').lower() == query_filter.lower():
                    filtered_sentences.append(evidence.get('evidence_extract_from_source', ''))
    return filtered_sentences

# Function to preprocess text
def preprocess_text(text):
    stop_words = set(stopwords.words('english'))
    tokens = word_tokenize(text)
    tokens = [word.lower() for word in tokens if word.isalpha() and word not in stop_words]
    return tokens

# Function to perform topic modeling and return the model and corpus
def topic_modeling(texts, num_topics=5, num_words=10):
    # Preprocess texts
    processed_texts = [preprocess_text(text) for text in texts]

    # Create a dictionary and corpus
    dictionary = Dictionary(processed_texts)
    corpus = [dictionary.doc2bow(text) for text in processed_texts]

    # Train LDA model
    lda = LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=15)

    return lda, corpus, dictionary

# Mount Google Drive
drive.mount('/content/drive')

# Replace 'path_to_your_directory' with the actual path
path = '/content/drive/My Drive/lobbymap_dataset'

# Replace 'your_sentence_level_jsonl_file.jsonl' with the actual file name
jsonl_file_path = os.path.join(path, 'train.jsonl')

# Define the query you want to filter by
# query_filter = "Alignment with IPCC on Climate Action" # Replace with your desired query
# query_filter = "Carbon Tax"
query_filter = "GHG Emission Regulation"


# Load and filter the JSONL file based on the query
data = load_jsonl(jsonl_file_path)
filtered_sentences = filter_data(data, query_filter)

# # Print filtered sentences for debugging
# print("\nFiltered Sentences:")
# for sentence in filtered_sentences[:5]:  # Print the first 5 for inspection
#     print(sentence)

# Perform topic modeling
print("\nPerforming Topic Modeling...")
lda_model, corpus, dictionary = topic_modeling(filtered_sentences)

# Perform topic modeling
print("\nTopic Modeling Results:")
topic_modeling(filtered_sentences)

# Visualize the topics using pyLDAvis
print("\nGenerating Visualization...")
vis_data = gensimvis.prepare(lda_model, corpus, dictionary)
pyLDAvis.display(vis_data)


# # Save the visualization as an HTML file locally
# local_html_file_path = '/content/topic_modeling_visualization.html'
# pyLDAvis.save_html(vis_data, local_html_file_path)
# print(f"Visualization saved as HTML to {local_html_file_path}")

# # Copy the HTML file to Google Drive
# drive_html_file_path = os.path.join(path, 'topic_modeling_visualization.html')
# shutil.copy(local_html_file_path, drive_html_file_path)
# print(f"Visualization copied to Google Drive at {drive_html_file_path}")
```
