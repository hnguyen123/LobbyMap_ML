```python
import json
from collections import defaultdict
import pandas as pd

def tabulate_data_by_year(input_file):
    data_by_year = defaultdict(lambda: {'unique_documents': set(), 'unique_sentences': set()})
    
    with open(input_file, 'r', encoding='utf-8') as infile:
        for line in infile:
            data = json.loads(line)
            document_id = data.get('document_id')
            sentences = data.get('sentences', [])
            meta = data.get('meta', {})
            meta_evidences = meta.get('evidences', [])
            
            for evidence_list in meta_evidences:
                for evidence in evidence_list:
                    evidence_year = evidence.get('evidence_year')
                    if evidence_year:
                        data_by_year[evidence_year]['unique_documents'].add(document_id)
                        for sentence in sentences:
                            data_by_year[evidence_year]['unique_sentences'].add((document_id, sentence['sentence_id']))
    
    summary_data = {
        'Year': [],
        'Unique Document IDs': [],
        'Unique Sentence IDs': []
    }
    
    for year, counts in sorted(data_by_year.items()):
        summary_data['Year'].append(year)
        summary_data['Unique Document IDs'].append(len(counts['unique_documents']))
        summary_data['Unique Sentence IDs'].append(len(counts['unique_sentences']))
    
    df = pd.DataFrame(summary_data)
    return df

# Define the path to your combined file
combined_file = 'data/processed/combined.jsonl'

# Generate the summary table
summary_table = tabulate_data_by_year(combined_file)

# Print the summary table
print(summary_table)


# Save the summary table to a CSV file
summary_table.to_csv('data/processed/summary_table.csv', index=False)
```

```python
import json
import pandas as pd
from collections import defaultdict

def list_document_ids_for_company_and_year(input_file, target_company, target_year):
    documents_info = defaultdict(set)
    all_seen_documents = set()
    
    with open(input_file, 'r', encoding='utf-8') as infile:
        for line in infile:
            data = json.loads(line)
            document_id = data.get('document_id')
            meta = data.get('meta', {})
            meta_evidences = meta.get('evidences', [])
            
            for evidence_list in meta_evidences:
                for evidence in evidence_list:
                    evidence_year = evidence.get('evidence_year')
                    evidence_region = evidence.get('evidence_region')
                    company_name = evidence.get('company_name')
                    
                    if company_name == target_company and evidence_year == target_year:
                        if document_id not in all_seen_documents:
                            documents_info[evidence_region].add(document_id)
                            all_seen_documents.add(document_id)
    
    return documents_info

# Define the path to your combined file
combined_file = 'data/processed/combined.jsonl'

# List all document_id and evidence_region for a specified company and year
target_company = 'AES Corporation'
target_year = 2022
documents_info = list_document_ids_for_company_and_year(combined_file, target_company, target_year)

# Convert the dictionary to a DataFrame
rows = []
for region, document_ids in documents_info.items():
    for doc_id in document_ids:
        rows.append({'evidence_region': region, 'document_id': doc_id})

df = pd.DataFrame(rows)

# Define the path to save the CSV file
csv_file_path = 'data/processed/unique_region_evidence.csv'

# Save the DataFrame to a CSV file
df.to_csv(csv_file_path, index=False)

# Display the DataFrame
print(df)
```

```python
import json
import pandas as pd
from collections import defaultdict

def find_documents_with_different_years(input_file):
    evidence_years_by_doc = defaultdict(set)
    
    with open(input_file, 'r', encoding='utf-8') as infile:
        for line in infile:
            data = json.loads(line)
            document_id = data.get('document_id')
            meta = data.get('meta', {})
            meta_evidences = meta.get('evidences', [])
            
            for evidence_list in meta_evidences:
                for evidence in evidence_list:
                    evidence_year = evidence.get('evidence_year')
                    if evidence_year:
                        evidence_years_by_doc[document_id].add(evidence_year)
    
    docs_with_multiple_years = {
        doc_id: list(years) for doc_id, years in evidence_years_by_doc.items() if len(years) > 1
    }
    
    return docs_with_multiple_years

# Define the path to your combined file
combined_file = 'data/processed/combined.jsonl'

# Find documents associated with different years
docs_with_different_years = find_documents_with_different_years(combined_file)

# Convert the results to a DataFrame
df = pd.DataFrame([
    {'document_id': doc_id, 'years': years}
    for doc_id, years in docs_with_different_years.items()
])

# Define the path to save the CSV file
csv_file_path = 'data/processed/documents_with_different_years.csv'

# Save the DataFrame to a CSV file
df.to_csv(csv_file_path, index=False)

# Print the DataFrame
print(df)

```
